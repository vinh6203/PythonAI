```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# Image Thresholding la 1 phuong phap co ban de phan vung anh (segmenting image)
# Tuy nhien se chi hoat dong duoc neu co it noise

filename = 'mountain.jpg'
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
plt.imshow(img)
plt.title('Original Image')
```

```python
# Simple Thresholding
# Bat dau voi mot anh (gray) ta se chuyen tung pixel = 1 (hoac 255) if pixel_intensity > threshold, else 0
# Khi tang threshold => so pixel thoa man giam dan => anh nhan duoc se co it pixel sang hon

f, arr = plt.subplots(1, 3)
arr[0].imshow(cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]) # threshold = 127
arr[0].set_title('Threshold = 127')
arr[0].set_xticks([]), arr[0].set_yticks([])

arr[1].imshow(cv2.threshold(img, 175, 255, cv2.THRESH_BINARY)[1]) # threshold = 175
arr[1].set_title('Threshold = 175')
arr[1].set_xticks([]), arr[1].set_yticks([])

arr[2].imshow(cv2.threshold(img, 225, 255, cv2.THRESH_BINARY)[1]) # threshold = 225
arr[2].set_title('Threshold = 225')
arr[2].set_xticks([]), arr[2].set_yticks([])
```

```python
# Trong thu vien cv2 cung ho tro Adaptive Mean va Gaussian Thresholding de xu ly cho viec anh co anh sang khong deu

# Adaptive Mean se quyet dinh viec 1 pixel > threshold bang cach lay trung binh trong cac pixel lan can va tru di (constant) C
# neu ket qua > 0 thi pixel = 1 (sang), else pixel = 0 (toi)

# Adaptive Gaussian se quyet dinh viec 1 pixel > threshold bang cach lay tong cac diem voi trong so tu phan phoi Gauss trong cac 
# pixel lan can va tru di (constant) C neu ket qua > 0 thi pixel = 1 (sang), else pixel = 0 (toi)
```

```python
# Trong thu vien cv2 cung ho tro Adaptive Mean va Gaussian Thresholding de xu ly cho viec anh co anh sang khong deu
filename = 'brightnessDiff.jpg'
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, simple_threshold = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
adaptive_threshold_mean = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2) # window size = 11, bias_C = 2
adaptive_threshold_gaussian = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

f2, arr2 = plt.subplots(2, 2)
arr2[0,0].imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
arr2[0,0].set_title('Original')
arr2[0,0].set_xticks([]), arr2[0, 0].set_yticks([])

arr2[0,1].imshow(cv2.cvtColor(simple_threshold, cv2.COLOR_GRAY2RGB))
arr2[0,1].set_title('Simple Threshold, th = 127')
arr2[0,1].set_xticks([]), arr2[0,1].set_yticks([])

arr2[1,0].imshow(cv2.cvtColor(adaptive_threshold_mean, cv2.COLOR_GRAY2RGB))
arr2[1,0].set_title('Adaptive Mean Threshold')
arr2[1,0].set_xticks([]), arr2[1,0].set_yticks([])

arr2[1,1].imshow(cv2.cvtColor(adaptive_threshold_gaussian, cv2.COLOR_GRAY2RGB))
arr2[1,1].set_title('Adaptive Gaussian Threshold')
arr2[1,1].set_xticks([]), arr2[1,1].set_yticks([])
```

```python
# Tu cai dat ham Simple thresholding
def simpleThresholding(img, threshold=127):
    return np.uint8( (img > threshold)*255 )

plt.imshow(cv2.cvtColor(simpleThresholding(img), cv2.COLOR_GRAY2RGB))
```
