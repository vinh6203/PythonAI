```python
from sklearn.ensemble import RandomForestRegressor
from boruta import BorutaPy
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
X = np.array(df[:8])
```

```python
X = pd.read_csv("diabetes.csv", usecols = ['Pregnancies','Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
X = np.array(X)
y = np.array(pd.read_csv("diabetes.csv", usecols=['Outcome']))
print(X)
print(y)
```

```python
model = RandomForestRegressor(n_estimators=100, max_depth=6, random_state=42)

feat_selector = BorutaPy(verbose = 2, estimator=model, n_estimators='auto', max_iter=20)

feat_selector.fit(X, y)
```

```python
print("\n------Support and Ranking for each feature------")
for i in range(len(feat_selector.support_)):
    if feat_selector.support_[i]:
        print("Passes the test: ", 0,
              " - Ranking: ", feat_selector.ranking_[i])
    else:
        print("Doesn't pass the test: ",
              0, " - Ranking: ", feat_selector.ranking_[i])
```

```python
df = pd.read_csv("diabetes.csv")

label = df['Outcome']
df = df.drop('Outcome',axis = 1)
data = df

from sklearn.model_selection import train_test_split

x_train1,x_test1,y_train1,y_test1 = train_test_split(data,label,test_size = 0.3)
```

```python
x_train1 = x_train1/255
x_test1 = x_test1/255
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNC

knn = KNC(n_neighbors = 12,weights = 'distance',p = 2)
knn.fit(x_train1,y_train1)
predictKNN = knn.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test1,predictKNN)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train1, y_train1)

y_pred = rf.predict(x_test1)

accuracy = accuracy_score(y_test1, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.svm import SVC
svc = SVC()

svc.fit(x_train1,y_train1)
predictSVC = svc.predict(x_test1)

accuracy_score(y_test1,predictSVC)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
model = Sequential([
	Dense(units = 128,activation = 'relu'),
    Dense(units = 32,activation = 'relu',kernel_regularizer=tf.keras.regularizers.l2(0.1)),
	Dense(units = 4,activation = 'relu'),
    Dense(units = 1,activation = 'sigmoid'),
	])
from tensorflow.keras.losses import BinaryCrossentropy
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001),
)
model.fit(x_train,y_train,epochs = 150)
```

```python
def predict01(test_set):
    predict = model.predict(test_set)
    for i in range (0,len(predict)):
        tmp = predict[i]
        if (tmp > 0.5): predict[i] = int(1)
        else: predict[i] = int(0)
    predict = np.asarray(predict,dtype = np.int32)
    return predict

predictMLP = predict01(x_train)
accuracy_score(predictMLP,y_train)
```
