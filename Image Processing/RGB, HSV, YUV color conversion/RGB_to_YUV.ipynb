```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
filename = 'cat.jpg'
yuvFilename = 'yuv_' + filename
img = cv2.imread(filename)  #BGR
img[:,:,[0,2]] = img[:,:,[2,0]] #BGR -> RGB
plt.imshow(img)
```

```python
print(type(img))
print(img.shape)
```

```python
A = np.array([[  .299 ,  .587  ,  .114],
              [ -.1687, -.3313 ,  .5  ],
              [  .5   , -.4187 , -.0813]]).T

def RGB2YUV(img):
    global A
    yuv = img.dot(A)
    yuv[:,:,[1,2]] += 128
    return np.uint8(yuv)
```

```python
yuvImage = RGB2YUV(img)
plt.imshow(yuvImage)
```

```python
cv2.imwrite(yuvFilename, yuvImage)
```
