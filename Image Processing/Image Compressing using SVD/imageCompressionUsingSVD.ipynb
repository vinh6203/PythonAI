```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('portrait.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

```python
u, s, v = np.linalg.svd(img)
k = 10
for i in range(1, 4):
    approx = np.matmul(u[:, :k] * s[:k], v[:k, :])
    
    plt.subplot(1, 3, i)
    plt.imshow(approx, cmap='gray')
    plt.title('Number of singular value %i' % (k))
    
    cv2.imwrite('sigma%i.jpg' % (k), approx) # Save image
    k += 50
```
